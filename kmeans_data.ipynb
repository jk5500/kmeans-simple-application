```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2] 
Y = iris.target
```

```python
k=3
color=['rx','go','bx']
for i in range(k):
    index=np.argwhere(Y==i)
    plt.plot(X[index,0],X[index,1],color[i])
    plt.grid()
```

```python
plt.plot(X[:,0],X[:,1],'bx')
plt.grid()
```

```python
k=3
mu=X[np.random.choice(X.shape[0],k,replace=False),:]
new=np.zeros(X.shape[0])
mu
```

```python
for it in range(50):
    for i in range(len(X)):
        distance=np.sqrt((X[i,0]-mu[:,0])**2+(X[i,1]-mu[:,1])**2)
        center=np.argwhere(distance==np.min(distance))
        new[i]=center[0]

    for f in range(k):
        r=new==f
        mu[f]=np.sum(X[r],axis=0)/np.sum(r)
    
    
    
mu=mu
```

```python

center_all=[]
for i in range(X.shape[0]):
    distance=np.sqrt((X[i,0]-mu[:,0])**2+(X[i,1]-mu[:,1])**2)
    center=np.argwhere(distance==np.min(distance))
    center_all.append(center[0])
center_all=np.array(center_all)   
k=3
color=['rx','go','bx']
for i in range(k):
    index=np.argwhere(center_all==i)
    plt.plot(X[index,0],X[index,1],color[i])
    
    
    
```
