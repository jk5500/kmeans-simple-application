```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, [1,2]]  # we only take the first two features.
Y = iris.target
plt.plot(X[:,0],X[:,1],'ro')
plt.title('iris data set')
plt.show()
```

```python
np.random.seed(0)
k=3
```

```python
center=X[np.random.choice(range(X.shape[0]),k,replace=False)]
color=['ro','go','bo']
energyf=[]

for iteration in range(5):
    labels=[]
    
    energy_initial=[]
    for n in range(X.shape[0]):
        distance=(X[n]-center)**2  
        dist_sum=np.sum(distance,axis=1)
        label=np.array(np.argmin(dist_sum))
        value_energy=np.sum(distance[label])
        energy_initial.append(value_energy)
        labels.append(label)

        
   
    labels=np.array(labels)
    energy_it = np.sum(energy_initial)

    c1=np.sum(X[labels==0],axis=0)/np.sum(labels==0)
    c2=np.sum(X[labels==1],axis=0)/np.sum(labels==1)
    c3=np.sum(X[labels==2],axis=0)/np.sum(labels==2)
    center=np.array([c1,c2,c3])

    energyf.append(energy_it)
    
    plt.figure(1)
    
    plt.plot(X[labels==0,0],X[labels==0,1],'ro')
    plt.plot(X[labels==1,0],X[labels==1,1],'go')
    plt.plot(X[labels==2,0],X[labels==2,1],'bo')
    plt.title(  str(iteration))
    plt.show()
    plt.pause(0.005)
```

```python
plt.plot(energyf)
```
